### 問1

文字列"datascience"の文字を逆に（末尾から先頭に向かって）並べた文字列を出力してください

```python
#スライシング
s = "datascience"
reverce_s = s[::-1]
print(reverce_s)

#reverced()関数とjoin()メソッドを使う
s = "datascience"
reverce_s = ''.join(reversed(s)) #joinメソッドを使わなければただリストが返される。joinメソッドで文字列に変換。''で区切り文字なしを指定。
print(reverce_s)

#forループを使う
s = "datascience"
reverce_s = ""
for char in s:
    reverce_s = char + reverce_s #charをreverce_sの後ろにすると順番が変わらない
print((reverce_s))
```

### 問3

「ヘルシンキ」＋「エストニア」の文字を先頭から交互に連結して文字列「ヘエルスシトンニキア」を出力してください


```python
#forループを使う
s = "ヘルシンキ"
t = "エストニア"
country = ""
for i in range(min(len(s), len(t))):#ここでs,tのようにおくとそのまま出力(タプルとして認識)。len()関数で文字列の長さを取得し、min()関数で短い方の長さを取得（短い方に合わせている）。
    country = country + s[i] + t[i] #for文の下にfor文を書くと内側ループを止めても外側ループは次に行くものの、内側ループは最初から始まる。brekeしなかったらtが全部連結されてしまう。
print(country)
```

### 問4

"Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成してください


```python
s = "Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"
w = s.split() #スペースで区切ってリスト化
result = []
for word in w:
    result.append(len(word)) #リストに単語の長さを追加
print(result)
```

### 問5

"It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."という文を単語に分解し，1, 2, 5, 6, 8, 9, 16番目の単語の先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）を格納した辞書を作成してください

```python
s = "It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."
w = s.split() #スペースで区切ってリスト化
print(len(w))
position = [1,2,5,6,8,9,16]

result = {}
for pos in position:
    word = w[pos-1] #リストのインデックスは0から始まるので-1する
    first_two = word[:2] #単語の最初の2文字を取得
    result[pos] = first_two

print(result)
```
